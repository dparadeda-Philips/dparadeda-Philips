```python
#READ JSON FROM S3 BUCKET
import boto3
import json
from munch import munchify 

s3 = boto3.resource('s3')
for bucket in s3.buckets.all():
    bucket_name = bucket.name
    
# for obj in bucket.objects.all():
#     obj = s3.Object(bucket_name, obj.key)
#     print ('Bucket Name:', bucket_name, 'File Name:', obj.key)   

var_list = []    

for obj in bucket.objects.all():    
    json_content = json.loads(obj.get()['Body'].read().decode('utf-8'))
    var_name = str(obj.key).split(".json")[0]
    if 'laceplus' in var_name:
        var_list.append(var_name)
#         globals()[f"{var_name}"] = json_content
print (var_list)
```

```python
dict_location_start = {}
dict_location_end = {}
for var_name in var_list:
    for value in munchify(globals()[var_name]).entry:
        try:
            dict_location_start[var_name] = value.resource.location[0].period.start
            dict_location_end[var_name] = value.resource.location[0].period.end
        except:
            pass

for key in dict_location_start:
    print (key, dict_location_start[key], dict_location_end[key])
```
